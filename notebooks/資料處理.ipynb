```python
import pandas as pd
import numpy as np
import os
import shutil
import warnings

warnings.filterwarnings('ignore')

folders = ['data/raw', 'data/processed', 'outputs/figures', 'outputs/tables', 'outputs/summary']
for folder in folders:
    os.makedirs(folder, exist_ok=True)

source_path = "統計2/YRBS_2007.csv"
standard_raw_path = "data/raw/YRBS_2007.csv"
output_path = "data/processed/yrbs_cleaned.csv"

if os.path.exists(source_path):
    shutil.move(source_path, standard_raw_path)
    print(f"File moved from '{source_path}' to '{standard_raw_path}'")
elif os.path.exists("YRBS_2007.csv"):
    shutil.move("YRBS_2007.csv", standard_raw_path)
    print(f"File moved to: {standard_raw_path}")
elif os.path.exists(standard_raw_path):
    print(f"File already exists in {standard_raw_path}, proceeding to processing.")
else:
    print("Error: YRBS_2007.csv not found.")

if os.path.exists(standard_raw_path):
    df_raw = pd.read_csv(standard_raw_path)
    df_raw.columns = df_raw.columns.str.strip() 
    
    target_sad = "SadOrHopeless"
    target_height = "HowTallAreYouWithoutShoesInMeters"
    
    df = df_raw[[target_sad, target_height]].copy()

    print(f"\n--- [Data Check] ---")
    print(f"Original missing value statistics:\n{df.isnull().sum()}")

    df['Sad_Binary'] = df[target_sad].map({1.0: 1, 2.0: 0})

    df_clean = df.dropna(subset=['Sad_Binary', target_height]).copy()

    df_clean.to_csv(output_path, index=False)
    
    print("-" * 40)
    print(f"Variable definition and coding completed")
    print(f"Final number of samples analyzed (n): {len(df_clean)}")
    print(f"1. Raw file location: {standard_raw_path}")
    print(f"2. Processed file location: {output_path}")
    print("-" * 40)
    
    print("\n--- Preview of cleaned data ---")
    print(df_clean[['Sad_Binary', target_height]].head())
```
